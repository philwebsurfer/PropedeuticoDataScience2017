```python
# Recibir el path de un archivo de imagen png y convertirlo en una matriz numérica que represente a la versión en blanco y negro de la imagen
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import time
from PIL import Image
# Se debe de escribir el path de la imagen guardado. El que aparece actualmente está en mi computadora
img = Image.open('C:\\Users\\Roberto\\Documents\\Maestria Ciencia de Datos\\GitPropDataScience\\PropedeuticoDataScience2017\\Alumnos\\Roberto Acevedo\\Screen_Shot_2012-05-24_at_11.48.39_AM_grande.png')
imggray = img.convert('LA')
imgmat = np.array(list(imggray.getdata(band=0)), float)
imgmat.shape = (imggray.size[1], imggray.size[0])
imgmat = np.matrix(imgmat)
plt.figure(figsize=(9,6))
plt.imshow(imgmat, cmap='gray');
```

```python
# Realizar y verificar la descomposición svd
U, sigma, V = np.linalg.svd(imgmat)
```

```python
# Usar la descomposición para dar una aproximación de grado k de la imagen
reconstimg = np.matrix(U[:, :1]) * np.diag(sigma[:1]) * np.matrix(V[:1, :])
plt.imshow(reconstimg, cmap='gray');
```

```python
for i in range(2, 4):
    reconstimg = np.matrix(U[:, :i]) * np.diag(sigma[:i]) * np.matrix(V[:i, :])
    plt.imshow(reconstimg, cmap='gray')
    title = "n = %s" % i
    plt.title(title)
    plt.show()
```

```python
# Para alguna imagen de su elección, elegir distintos valores de aproximación a la imagen original.
for i in range(5, 51, 5):
    reconstimg = np.matrix(U[:, :i]) * np.diag(sigma[:i]) * np.matrix(V[:i, :])
    plt.imshow(reconstimg, cmap='gray')
    title = "n = %s" % i
    plt.title(title)
    plt.show()
```

```python
# Programar una función que dada cualquier matriz devuelva la pseudainversa usando la descomposición SVD. Hacer otra función que resuelva cualquier sistema de ecuaciones de la forma Ax=b usando esta pseudoinversa.
import numpy as np
from numpy.linalg import pinv
from numpy import transpose, dot
def calcpseudoinv(Array):
    MtxPsInv = pinv(Array)
    return MtxPsInv
def solucion(Array,b):
    nrenArray = len(Array)
    nrenb = len(b)
    if nrenArray != nrenb:
        raise Exception("El sistema no tiene solución")
    U, sigma, V = np.linalg.svd(Array)
    MtxPsInv = pinv(Array)
    Vt = V.transpose()
    Ut = U.transpose()
    solucion = dot(Vt,dot(MtxPsInv,dot(Ut,b)))
    return solucion
```
