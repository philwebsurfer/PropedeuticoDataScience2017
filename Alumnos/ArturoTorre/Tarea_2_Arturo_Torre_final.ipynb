## **1.- ¿Por qué una matriz equivale a una transformación lineal entre espacios vectoriales?**

### *R= Cuando una matriz $A$ multiplica a un vector $v$, lo transforma en otro vector dado $Av$. Ampliando dicha idea inicial se puede hablar de un conjunto de vectores $v$ que pertenecen al espacio $V$ que al ser multiplicados, cada uno por $A$, son transformados. Existen ocasiones en que contamos con el producto de las transformaciones lineales de los espacios vectoriales, en este caso $V$, pero desconocemos la matriz $A$. En ese sentido, podemos asignar una matriz a cada transformación. Para vectores columna la entrada $v$ pertenece a $V=R^n$ y la salida $T(v)$ pertenece a $Z=R^m$.   En ese sentido, la matriz $A$ para la transformación $A$, será de $m x n$. La selección de  bases en $V$ y $W$ determinarán la matriz $A$. Los vectores base estándar  de $R^n$ y $R^m$  son las columnas de la matriz identidad. Si escogemos dichos vectores tendremos una transformación de la forma $T(v)=Av$. Al mismo tiempo dichos espacios tienen otras bases posibles en la misma transformación representada por otras matrices.*                                  

### *Cuando $\textit{V}$ y $\textit{W}$ no son $R^n$ y $R^m$, siguen conteniendo bases. Cada elección determinada de base nos llevará a una matriz $\textit{T}$. Cuando la base de entrada es diferente de la base de salida, la matriz para $T(v)=v$ no será la matriz identidad. En lugar de eso será la matriz que cambia las bases.*        

### *Cada $\textit{v}$ es una combinación única $c_1 v_1+c_2 v_2+⋯+c_n$ $v_n$ de la base de vectores de $v_i$. Si T es una transformación linear, $T(v)$ debe ser la misma combinación  $c_1 T(v_1)+c_2 T(v_2)+⋯+c_n T(v_n)$ de las salidas conocidas $T(v_i)$. *

### *Es decir una transformación representa el simil de una función. En ese sentido existe un regla de que asocia los valores de un conjunto dado (dominio para el caso de funciones) con otro conjunto (codominio para el caso de funciones también). Específicamente, de lo explicado anteriormente, existe una matriz que nos ayuda a relacionar ambos espacios vectoriales. Esa matriz, es la matriz $\textit{A}$ del planteamiento inicial en la respuesta.*

$$T:R^n→R^m$$
$$T(v ⃗ ):Av ⃗$$

## Cálculo de la Pseudoinversa

```python
### Primero generamos una matriz aleatoria de números enteros con el siguiente código
A = np.random.randint(7, size=(4, 3))
```

```python
A
```

```python
import numpy as np
from numpy.linalg import inv
```

```python
def solucion(A):
    u, s, v = np.linalg.svd(A,0)
    B = np.dot(v.transpose(),np.dot(np.diag(s**-1),u.transpose()))
    return np.array(B)
```

```python
#Calcululamos la pseudoinversa de A
B = np.array(solucion (A))
B
```

### * Una de las propiedades de las matrices invertibles es que la inversa de la inversa da la matriz original.* 

$$ (A^{-1})^{-1}= A $$

```python
C = np.array(solucion (B))
C
```

```python
t_ = []
c_ = []
t_2 = []
c_2 = []
etas_ = []
costos = []

for i in range(len(etas)):
    (t_2, c_2) = gradiente2(x_dg,y_dg,teta,etas[i],iterar)
    t_.append(t_2) #guardo la alfa y la beta de cada eta
    c_.append(c_2) #guardo los errores de cada iteración * eta
    etas_.append(etas[i]) #voy guardando las etas
    costos.append(sum(c_2)) #quería obtener un error total de cada eta, lo sumé, son los errores de cada eta
```
